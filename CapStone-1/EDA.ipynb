```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("adult.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python

df.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

# Histogram of Age



```python
plt.hist(df['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('Distribution of age in the dataset')
plt.show()
```

#Distribution of Workclass



```python
plt.bar(df['workclass'].unique(), df['workclass'].value_counts())
plt.xlabel('Workclass')
plt.ylabel('Number of people')
plt.title('Distribution of workclass in the dataset')
plt.show()
```

```python
plt.pie(df['education'].value_counts(), autopct="%1.1f%%")
plt.title('Distribution of education in the dataset')
plt.show()
```

```python
plt.bar(df['marital_status'].unique(), df['marital_status'].value_counts())
plt.xlabel('Marital Status')
plt.ylabel('Number of people')
plt.title('Distribution of marital status in the dataset')
plt.show()
```

```python
plt.bar(df['occupation'].unique(), df['occupation'].value_counts())
plt.xlabel('Occupation')
plt.ylabel('Number of people')
plt.title('Distribution of occupation in the dataset')
plt.show()
```

```python
plt.bar(df['relationship'].unique(), df['relationship'].value_counts())
plt.xlabel('Relationship')
plt.ylabel('Number of people')
plt.title('Distribution of relationship in the dataset')
plt.show()
```

```python
plt.pie(df['race'].value_counts(), autopct="%1.1f%%")
plt.title('Distribution of race in the dataset')
plt.show()
```

```python
plt.bar(df['gender'].unique(), df['gender'].value_counts())
plt.xlabel('Gender')
plt.ylabel('Number of people')
plt.title('Distribution of gender in the dataset')
plt.show()
```
